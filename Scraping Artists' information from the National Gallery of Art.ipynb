```python
import requests
from bs4 import BeautifulSoup

page = requests.get('https://web.archive.org/web/20121007172955/http://www.nga.gov/collection/anZ1.htm')
soup = BeautifulSoup(page.text, 'html.parser') # The 'page.text' document from Requests is the content of the server's response
```

```python
# Pulling all text from the BodyText div
artist_name_list = soup.find(class_ = 'BodyText')

# Pulling text from all instances of the <a> tag within BodyText div
artist_name_list_items = artist_name_list.find_all('a')
```

```python
print (type(artist_name_list))
print (type(artist_name_list_items))
```

```python
# DIsplaying out the names of all the artists

for artist_name in artist_name_list_items:
    print (artist_name.prettify())
```

```python
# DIsplaying out the names of all the artists, not having used prettify()

for artist_name in artist_name_list_items:
    print (artist_name)
```

```python
# Removing the unnecessary bottom links
last_links = soup.find(class_ = 'AlphaNav')
last_links.decompose()

# Re-iterating
artist_name_list = soup.find(class_ = 'BodyText')
artist_name_list_items = artist_name_list.find_all('a')

for artist_name in artist_name_list_items:
    print (artist_name.prettify())
```

```python
# Using .contents for pulling out the <a> tag's children

for artist_name in artist_name_list_items:
    names = artist_name.contents[0]
    print (names)
```

```python
for artist_name in artist_name_list_items:
    print (artist_name.contents)
    
print ('\n', type(artist_name.contents))
```

```python
# For getting the URLs

for artist_name in artist_name_list_items:
    names = artist_name.contents[0]
    links = 'https://web.archive.org' + artist_name.get('href') # Since as observed in above, the entire URL isn't getting captured
    print (names)
    print (links)
```

```python
# Doing for a single page

import csv

artist_name_list = soup.find(class_ = 'BodyText')
artist_name_list_items = artist_name_list.find_all('a')

with open('artists.csv', 'w', newline = '') as csvfile:
    fieldnames = ['Name', 'Link']
    writer = csv.DictWriter(csvfile, fieldnames = fieldnames)
    writer.writeheader()
    for artist_name in artist_name_list_items:
        names = artist_name.contents[0]
        links = 'https://web.archive.org' + artist_name.get('href')
        writer.writerow({'Name' : names, 'Link' : links})
```

```python
# Now, we use all the pages

pages = []

for i in range(1, 5):
    url = 'https://web.archive.org/web/20121007172955/https://www.nga.gov/collection/anZ' + str(i) + '.htm'
    pages.append(url)

with open('artists.csv', 'w', newline = '') as csvfile:
    fieldnames = ['Name', 'Link']
    writer = csv.DictWriter(csvfile, fieldnames = fieldnames)
    writer.writeheader()
    for item in pages:
        page = requests.get(item)
        soup = BeautifulSoup(page.text, 'html.parser')
        last_links = soup.find(class_ = 'AlphaNav')
        last_links.decompose()
        artist_name_list = soup.find(class_ = 'BodyText')
        artist_name_list_items = artist_name_list.find_all('a')
        for artist_name in artist_name_list_items:
            names = artist_name.contents[0]
            links = 'https://web.archive.org' + artist_name.get('href')
            writer.writerow({'Name' : names, 'Link' : links})        
        print (item)    
```
